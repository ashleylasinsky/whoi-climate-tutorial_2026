### Option 1: SST in future CESM scenario

`````python
# Path to the data
server_path = '/Volumes/cmip6'
path_on_server = '/data/cmip6/ScenarioMIP/NCAR/CESM2/ssp585/r1i1p1f1/Omon/tos/gr/1'
path = pathlib.Path(server_path + path_on_server)

# Get list of files
files = list(path.glob("*.nc"))

# Open dataset using xarray
ds = xr.open_mfdataset(files)
`````

### Option 2: Chlorophyll in historical GFDL-ESM4 Simulation

`````python
# Path to the data
server_path = '/Volumes/cmip6'
path_on_server = '/data/cmip6/CMIP/NOAA-GFDL/GFDL-ESM4/historical/r1i1p1f1/Omon/chl/gr/1'
path = pathlib.Path(server_path + path_on_server)

# Get list of files
files = list(path.glob("*.nc"))[-3:]

# Open dataset using xarray
ds = xr.open_mfdataset(files)
`````

### Option 3: Precipitation in ERA5 Reanalysis

`````python
# Path to the data
server_path = '/Volumes/cmip6'
path_on_server = '/data/era5/reanalysis/single-levels/monthly-means/total_precipitation'
path = pathlib.Path(server_path + path_on_server)

# Get list of files
files = list(path.glob("*.nc"))[0:-5]

# Open dataset using xarray
ds = xr.open_mfdataset(files)
`````

### Option 4: Ocean Current Velocity in ECCO Ocean Reanalysis

`````python
# U velocity
path = pathlib.Path('/Volumes/cmip6/data/ocean_reanalysis/ECCOv5/Version5/Alpha/latlon_monthly/UVEL.nc')
files = list(path.glob("*.nc"))
U_ds = xr.open_mfdataset(files[0:12], decode_times=False).UVEL

# V velocity
path = pathlib.Path('/Volumes/cmip6/data/ocean_reanalysis/ECCOv5/Version5/Alpha/latlon_monthly/VVEL.nc')
files = list(path.glob("*.nc"))
V_ds = xr.open_mfdataset(files[0:12], decode_times=False).VVEL

# Manually decode TIME
start_date = pd.Timestamp("1991-12-01")
months_since = U_ds["TIME"].values
U_ds['TIME'] = [start_date + pd.DateOffset(months=int(m)) for m in months_since]
V_ds['TIME'] = [start_date + pd.DateOffset(months=int(m)) for m in months_since]

# Combine U and V variables into one dataset using xarray
ds = xr.merge([U_ds, V_ds])
`````
